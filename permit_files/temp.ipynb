```python
# COUNT - testing

import csv
import re

# Function to query the dataset based on user input
def query_permit_data(dataset_path, user_input):
    # Read the CSV dataset
    data = []
    with open(dataset_path, 'r', newline='') as csvfile:
        csvreader = csv.DictReader(csvfile)
        for row in csvreader:
            data.append(row)

    # Split the user input into query components
    query_components = re.split(r'\s+', user_input)
    select_clause = query_components[1:query_components.index('from')]
    from_clause = query_components[query_components.index('from') + 1]
    where_index = query_components.index('where') if 'where' in query_components else None

    # Check if COUNT is in the SELECT clause
    count_field = None
    if 'count(' in user_input:
        count_match = re.search(r'count\((.*?)\)', user_input)
        if count_match:
            count_field = count_match.group(1)

    # Filter the dataset based on the WHERE clause if present
    filtered_data = data
    if where_index is not None:
        where_clause = ' '.join(query_components[where_index + 1:])
        filtered_data = [row for row in data if evaluate_condition(row, where_clause)]

    # Extract the SELECTed columns from the filtered data, including COUNT
    result = []
    for row in filtered_data:
        selected_data = {}
        for column in select_clause:
            if column.lower().startswith('count(') and count_field:
                # Count the occurrences of the specified field
                field_to_count = count_field.lower()
                count = sum(1 for row in filtered_data if row[field_to_count] == row[column])
                selected_data[column] = count
            else:
                selected_data[column] = row[column]
        result.append(selected_data)

    return result

# Function to evaluate the WHERE condition
def evaluate_condition(row, condition):
    operators = ['>', '<', '>=', '<=', '=', '!=']
    for operator in operators:
        if operator in condition:
            column, value = condition.split(operator)
            column = column.strip()
            value = value.strip()

            if operator == '=':
                operator = '=='

            if '*' in value:
                # If the value contains "*", perform wildcard matching
                value_pattern = value.replace('*', '.*')
                if re.match(value_pattern, row.get(column)) is not None:
                    return True
            else:
                # Perform standard comparison
                if column in row and eval(f'"{row[column]}" {operator} "{value}"'):
                    return True
    return False

# Example usage
dataset_path = 'xpermit_data.csv'
user_input = "select primary_scope_code count(record_id) from data where valuation > 10000"
results = query_permit_data(dataset_path, user_input)

for i, result in enumerate(results):
    for column, value in result.items():
        print(f'{column}: {value}')
    # Add a separator between records except for the last one
    if i < len(results) - 1:
        print('-' * 20)  # Use any separator you prefer
```

```python
#  using pandas

import pandas as pd

def execute_sql_query(data, query):
    # Split the query into SELECT and WHERE parts
    select_query, where_query = query.lower().split("where", 1)
    
    # Extract the columns to be selected
    select_columns = [col.strip() for col in select_query.split("select")[1].strip().split(",")]
    
    # Execute the query and apply filters
    result = data.query(where_query.strip())[select_columns]

    return result

# Sample data
data = pd.read_csv("xpermit_data.csv")

# User input query
user_query = "select record_id, open_date, contractor_phone, last_updated from data where parcel_number == '185-332-09-66'"
# user_query = "select * from data"

# Execute the query
result = execute_sql_query(data, user_query)
headers = data.columns
print(headers)
print(result)
```
